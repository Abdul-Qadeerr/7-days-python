# ‚úÖ To-Do List Manager in Python
A simple project to manage tasks.

```python
tasks = []  # List to store tasks

def add_task():
    task = input("Enter new task: ")
    tasks.append({"task": task, "completed": False})
    print("‚úÖ Task added!")

def view_tasks():
    if not tasks:
        print("‚ö†Ô∏è No tasks found!")
    else:
        print("\n--- To-Do List ---")
        for i, t in enumerate(tasks, 1):
            status = "‚úî Completed" if t['completed'] else "‚ùå Pending"
            print(f"{i}. {t['task']} [{status}]")

def update_task():
    view_tasks()
    if tasks:
        choice = int(input("Enter task number to update: "))
        if 1 <= choice <= len(tasks):
            tasks[choice-1]['task'] = input("Enter new task description: ")
            print("‚úèÔ∏è Task updated!")
        else:
            print("‚ö†Ô∏è Invalid task number!")

def delete_task():
    view_tasks()
    if tasks:
        choice = int(input("Enter task number to delete: "))
        if 1 <= choice <= len(tasks):
            tasks.pop(choice-1)
            print("üóë Task deleted!")
        else:
            print("‚ö†Ô∏è Invalid task number!")

def mark_completed():
    view_tasks()
    if tasks:
        choice = int(input("Enter task number to mark completed: "))
        if 1 <= choice <= len(tasks):
            tasks[choice-1]['completed'] = True
            print("‚úî Task marked as completed!")
        else:
            print("‚ö†Ô∏è Invalid task number!")

def mark_incomplete():
    view_tasks()
    if tasks:
        choice = int(input("Enter task number to mark incomplete: "))
        if 1 <= choice <= len(tasks):
            tasks[choice-1]['completed'] = False
            print("‚ùå Task marked as incomplete!")
        else:
            print("‚ö†Ô∏è Invalid task number!")

def menu():
    while True:
        print("\n===== To-Do List Manager =====")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Update Task")
        print("4. Delete Task")
        print("5. Mark Completed")
        print("6. Mark Incomplete")
        print("0. Exit")

        choice = input("Enter choice: ")

        if choice == '1':
            add_task()
        elif choice == '2':
            view_tasks()
        elif choice == '3':
            update_task()
        elif choice == '4':
            delete_task()
        elif choice == '5':
            mark_completed()
        elif choice == '6':
            mark_incomplete()
        elif choice == '0':
            print("üëã Exiting To-Do List Manager. Goodbye!")
            break
        else:
            print("‚ö†Ô∏è Invalid choice! Try again.")

# Run the menu
menu()
```
