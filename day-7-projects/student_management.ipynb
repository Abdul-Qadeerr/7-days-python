# ğŸ“ Student Management System
This is a simple project in Python using lists & dictionaries.

```python
students = []  # list to store student records

def add_student():
    roll = input("Enter Roll No: ")
    name = input("Enter Name: ")
    age = input("Enter Age: ")
    grade = input("Enter Grade: ")
    student = {"roll": roll, "name": name, "age": age, "grade": grade}
    students.append(student)
    print("âœ… Student added successfully!")

def view_students():
    if not students:
        print("âš ï¸ No students found!")
    else:
        print("\n--- Student List ---")
        for s in students:
            print(f"Roll: {s['roll']}, Name: {s['name']}, Age: {s['age']}, Grade: {s['grade']}")

def search_student():
    roll = input("Enter Roll No to search: ")
    for s in students:
        if s['roll'] == roll:
            print(f"Found â†’ Roll: {s['roll']}, Name: {s['name']}, Age: {s['age']}, Grade: {s['grade']}")
            return
    print("âŒ Student not found!")

def update_student():
    roll = input("Enter Roll No to update: ")
    for s in students:
        if s['roll'] == roll:
            s['name'] = input("Enter New Name: ")
            s['age'] = input("Enter New Age: ")
            s['grade'] = input("Enter New Grade: ")
            print("âœ… Student updated successfully!")
            return
    print("âŒ Student not found!")

def delete_student():
    roll = input("Enter Roll No to delete: ")
    for s in students:
        if s['roll'] == roll:
            students.remove(s)
            print("ğŸ—‘ Student deleted successfully!")
            return
    print("âŒ Student not found!")

def menu():
    while True:
        print("\n===== Student Management System =====")
        print("1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Update Student")
        print("5. Delete Student")
        print("0. Exit")
        
        choice = input("Enter choice: ")
        
        if choice == '1':
            add_student()
        elif choice == '2':
            view_students()
        elif choice == '3':
            search_student()
        elif choice == '4':
            update_student()
        elif choice == '5':
            delete_student()
        elif choice == '0':
            print("ğŸ‘‹ Exiting program. Goodbye!")
            break
        else:
            print("âš ï¸ Invalid choice! Try again.")

# Run the menu
menu()
```
